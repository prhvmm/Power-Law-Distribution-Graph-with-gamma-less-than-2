```python
import networkx as nx
import random as rand
import matplotlib.pyplot as plt
import numpy as np
```

#Creating the graph

```python

gamma = 1.5

n = 500

degree_seq = np.random.zipf(gamma, n)

if sum(degree_seq) % 2 != 0: #Making sure that the number of edges are even
    degree_seq[np.random.randint(n)] += 1

graph = nx.configuration_model(degree_seq)

graph.remove_edges_from(nx.selfloop_edges(graph)) #Removing self-loops

graph = nx.Graph(graph) #Removing multiple links

```

#log-log Degree Distribution

```python
degree_freq = nx.degree_histogram(graph)

# Compute pk for each degree
n_nodes = len(graph.nodes())
pk = [freq / n_nodes for freq in degree_freq]

# Plot the degree distribution in a log-log scale
degrees = range(len(degree_freq))
plt.loglog(degrees, pk, 'ro')
plt.xlabel('Degree (log scale)')
plt.ylabel('p_k (log scale)')
plt.title('Degree Distribution with 1000 nodes')
plt.show()







```
